```python
## --- venns of supplementary and our running results

# serfish
figure, axes = plt.subplots(7,3, figsize=(15,15), constrained_layout=True)   
for i in np.arange(33,53+1):
    #i=34
    df_s='supplementary_results/field_'+str(i)+'_sig_gene_list.csv'
    if os.path.exists(df_s):
        df_supply=pd.read_csv(df_s)
        sig_list_supply=df_supply.sig_gene
    else:
        sig_list_supply=pd.Series()
        
    df_r='running_results/field_'+str(i)+'sig_list.csv'
    if os.path.exists(df_r):
        df_run=pd.read_csv(df_r)
        sig_list_run=df_run.sig_list
    else:
        sig_list_run=pd.Series()
     
    x = int((i-33)/3)
    y= (i-33) % 3
    
    titleText = 'Field ' + str(i)
    venn2(subsets=[set(sig_list_supply.values),
                  set(sig_list_run.values)], 
                    set_labels = ('supplymentary','run_script'),                     
                   ax=axes[x][y])
    ax = axes[x, y]
    ax.axis('off')
    ax.set_title(titleText, fontname="Arial", fontsize=14)
axes[3,1].axis('off')
axes[3,2].axis('off')
#axes[3,3].axis('off')

plt.savefig('Comapre_results/SeqFish_supply_run.scrip.pdf')
plt.show()
```
