```python
import pandas as pd
```

```python
exp_mat=pd.read_excel('Data/Serfish_hippo/mmc6.xlsx',0,header=None,index_col=0)
exp_mat.iloc[:8,:20]
```

```python
exp_mat.index=exp_mat.index.map(eval)  ## remove ''
exp_mat.iloc[:8,:20]
```

```python

info=pd.read_excel('Data/Serfish_hippo/mmc6.xlsx',2,header=None)
info.index+=1
info.columns=['x','y','z']
info.shape
```

```python
filter_info=info[(info['x']<822)&(info['x']>203)&(info['y']<822)&(info['y']>203)]

# for scale some part as the paper's method of 'Trendsceek'  
```

```python
filter_info.shape
```

```python
filter_info.iloc[:10,]
```

```python
corr=[str(filter_info['x'].values[i])+'x'+str(filter_info['y'].values[i]) for i in range(len(filter_info))]
```

```python
filter_info.to_csv('Data/Serfish_hippo/serfish_info_2050.csv',index=0)
```

```python
exp_mat1=exp_mat.T.loc[filter_info.index]
exp_mat1.iloc[:10,:10]
```

```python
exp_mat1.insert(0,'info',corr)
exp_mat1.iloc[:10,:10]
```

```python
exp_mat1.shape
```

```python
exp_mat1.to_csv('Data/Serfish_hippo/serfish_counts_2050.csv')
```

```python
field_vector=pd.read_excel('Data/Serfish_hippo/mmc6.xlsx',1,header=None)
```

```python
field_vector=field_vector.T
field_vector.index+=1
field_vector.columns=['field']
```

```python
field_vector.shape
```

```python
field_vector1=field_vector.loc[filter_info.index]
```

```python
field_vector1.iloc[:10,]
```

```python
field_vector1.shape
```

```python
sample_info=filter_info.join(field_vector1)
sample_info['total_count']=exp_mat1.sum(axis=1).values
```

```python
sample_info.shape
```

```python
sample_info.iloc[:10,]
```

```python
sample_info.to_csv('Data/Serfish_hippo/sample_info_2050.cav',index=0)
```

```python
for field, field_si in sample_info.groupby('field'):
    field_si.to_csv(f'Data/Serfish_hippo/field_{field}_sample_info_0.csv')
    exp_mat1.loc[field_si.index].to_csv(f'Data/Serfish_hippo/field_{field}_counts_0.csv')
```
